```python
# Making a basic scatterplot with matplotlib

np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50) #random integers that range from 0-100 and we want 50 y-values to be generated.

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
#modifying the graph: making the scatterplot a bar chart 

fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
#horizontal bars instead
 
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y)
```

```python
#now making a line plot

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
#how do we make a histogram? In a histogram, we only want to y-values.

fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
# now we're going to add plot labels to the line graph.

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
# Now we're going to customize/modify the labels and titles.
# first, make a dictionary with properties we want for the labels.

font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

# now we assign these font dictionaries to the graphs.
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict=font1, loc='left') #loc arguments aligns it to wherever
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
#Remake our basic scatterplot, this time with colour and marker arguments

fig, ax = plt.subplots(figsize = (5,3))

ax.scatter(x, y,
           marker= "d",  #for more customization: https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers
           color = 'indigo')
```

```python
# Adjusting style and width of line connecting points

fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x, y,
           marker= '*', 
           color = 'indigo',  
           linestyle = '-',
           linewidth = 2)
```

```python
#modifying data points: colour

fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x, y,
           marker= '*', 
           color = '#7425b9',  
           linestyle = '-',
           linewidth = 2)
```

```python
# Further customization: markerzie, edgecolor and face color

fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x, y,
           marker= '*', 
           markersize = 12,     # new additino: size of the star
           color = '#7425b9',  
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359',     # new addition  (border)
           markerfacecolor = '#000000')     # new addition  (fill)
```

```python
# New addition: adding gridlines to the graph we already have

fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x, y,
           marker= '*', 
           markersize = 12,     
           color = '#7425b9',  
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359',     
           markerfacecolor = '#000000')     

ax.grid(                 # new addition  (gridlines and customization))
    axis = 'y',
    color = 'blue',
    linewidth = 2
    linestyle = '-.'
)
```
