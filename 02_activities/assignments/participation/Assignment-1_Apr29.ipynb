```python
#load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
import requests

#sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
# line plot with both y variables on same axes

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1)
ax.plot(x,y2)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()

# To add a legend, we add a 'label' argument to each plot object

ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right')
```

```python
#changing different elements of the legend

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = 'Person 1')
ax.plot(x,y2, label = 'Person 2')

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()

# To add a legend, we add a 'label' argument to each plot object

ax.legend(loc='lower right',
          frameon = True, #to add frame around the legend
          fontsize = 12, #change font size
          ncol = 2, #specify number of columns
          shadow = True)
```

```python
#Moving the legend outside of the plot area

#Specify the location to anchor the legend by using bbox_to_anchor

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
          bbox_to_anchor = (1,1))
```

```python
#Annotations with the text() function

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, 'This value is important!') #this text() function specifices coordinates and text
```

```python
# positioning text annotations with transforms

fig, ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5,", Data: (,5)", transform = ax.transData)
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure:(0.2,0.2)", transform=fig.transFigure)
```

```python
# Now annotating with arrows, using annotate() to annotate with arrows and text

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10,95), xytext=(20,94),
            arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
# Arrows can be modified

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10,95), xytext=(20,94),
            arrowprops=dict(arrowstyle='wedge', color='hotpink'))
```

```python
# Removing tick marks and labels by using locator and formatter

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Limiting tick marks 

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# Tick mark intervals

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Tick mark rotations (e.g. 45 degree rotation of the x axis labels)

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right')
```

```python
# modifying axis title fonts

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

font1={'family':'serif', 'color':'indigo'}
plt.xlabel('Shiny new X Axis!', fontsize=18, fontdict = font1)
```
