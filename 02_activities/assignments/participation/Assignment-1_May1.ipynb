```python
#load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
import requests

#sample data
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp","Sanji"])
y2 = np.array([110,180,240, 99, 220])
```

```python
# plot subplots: one scatter; one bargraph

fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
# customizing the above subplots: adding colour and titles for each subplot

fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7,3))
ax1.scatter(x1,y1, color='pink')
ax2.bar(x2,y2, color='green')

ax1.set_title('Scatter plot')
ax2.set_title('Bar graph')
```

```python
#subplots without a grid arrangement

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                     ['ax2', 'ax3']],
                                     figsize=(7,4))   

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
```

```python
#using layouts to make sure that subplots fit neatly within figure area

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                     ['ax2', 'ax3']],
                                     figsize=(7,4),
                                     layout = 'constrained')   

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A big label', fontsize=18)
someaxes["ax2"].set_xlabel('Another label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
```

```python
#Multiple ciz on one axes object

# sample data

x = np.array(["Luffy", "Zoro", "Nami", "Usopp","Sanji"])
y1 = np.array([110, 180, 240, 99, 240])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize = (7,3))

ax.bar(x,y1, color='indigo')
ax.plot(x,y2, color='red')


```

```python
# adding error information

y2_sd = np.std(y2)

# plot line as before

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = "red")

# add error information through errorbar()

ax.errorbar(x, y2, 
            yerr=y2_sd, 
            fmt="none",
            ecolor= 'indigo',
            elinewidth=4,
            capsize = 6,
            capthick = 2) #if we want to skip every other point, use errorevery = 2/3/4

#customizing errorbar appearance
```

```python
# adding images to plots

from PIL import Image # to open images
from io import BytesIO # to store images
```

```python
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content) #taking the contents of response and save it in a file
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color="red")

ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])

ax_image.imshow(image)
ax_image.axis('off')
```
